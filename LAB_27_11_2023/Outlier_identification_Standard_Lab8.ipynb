# tracer la courbe boite Ã  moustache en utilisant seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import seed, randn

# seed the random number generator
seed(1)

# generate univariate observations
data = 5 * randn(10000) + 50

# create a boxplot using Seaborn
sns.set(style="whitegrid")  # optional: set the style

# Plotting the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=data)

# Show the plot
plt.title('Boxplot')
plt.show()
```

# calculate summary statistics


```python
data_mean=mean(data)
data_std=std(data)
print(' mean=%.3f, stdv=%.3f' % (mean(data), std(data)))
```


# Calculer le seuil pour identifier les outliers Ã  plus de 3 Ã©carts-types de la moyenne.


```python

# Calcul de l'Ã©cart type
std_dev_value = np.std(data)

# Calcul de cut_off, lower et upper
cut_off = 3 * std_dev_value
lower, upper = np.mean(data) - cut_off, np.mean(data) + cut_off

print('lower=%0.3f, upper=%0.3f' % (lower, upper))
```

# identifier les valeurs aberrantes Ã  plus de 3 Ã©carts-types de la moyenne.

```python
outliers = np.random.uniform(low=80, high=120, size=160)

print('Identified outliers: %d' % len(outliers))
```


# remove outliers


```python
# Supprimer les outliers

outliers = (data < lower) | (data > upper)
outliers_removed = data[~outliers]

# Afficher le nombre d'observations non aberrantes
print('Non-outlier observations:', len(outliers_removed))
```
